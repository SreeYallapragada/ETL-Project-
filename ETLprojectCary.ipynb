```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Store Part I results into DataFrame
#Load the data to a DataFrame
cary_data = pd.read_csv("/Users/erikaberry/Desktop/Traffic_Accidents/data/cpd-crash-incidents.csv", encoding="utf-8")
cary_data.head() 
```

```python
is_2019 =  cary_data['year']==2019
print(is_2019.head())
```

```python
cary_accident_data_revised = cary_data[is_2019]
print(cary_accident_data_revised.shape)
```

```python
cary_accident_data_revised.round({"lat":6, "lon":6}) 
```

```python
# Filters and rearranges dataset to display most useful columns
cary_accident_data_revised_columns = cary_accident_data_revised[["TA_Date", "TA_Time", "lat", "lon", 
                                               "Weather", "Light_Condition"]]

cary_accident_data_revised_columns.head(20)
```

```python
# Rename the column headers
cary_data_to_compare = cary_accident_data_revised_columns.rename(columns={"TA_Date": "Date", "TA_Time": "Time", "lat": "Lat", 
                                              "lon": "Lng", "Light_Condition": "Nautical_Twilight"})

cary_data_to_compare.head(20)
```

```python
cary_data_to_compare.head()
```

```python
cary_data_to_compare.dropna(axis=0, how="any", thresh=None, subset=None, inplace=True)
cary_data_to_compare.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the head
cary_data_to_compare.to_csv("Cary_Traffic.csv", sep='|', index=False, header=True)
```
